```python
import numpy as np
import os

#  .npz 
npz_directory = '/home/hh/Documents/DeepLSD/my_experiment'

#  .npz 
npz_files = [f for f in os.listdir(npz_directory) if f.endswith('.npz')]

#  .npz 
for npz_file in npz_files:
    npz_path = os.path.join(npz_directory, npz_file)
    data = np.load(npz_path)
    
    print(f"Contents of {npz_file}:")
    for key in data.files:
        print(f" - {key}: shape {data[key].shape}")
```
