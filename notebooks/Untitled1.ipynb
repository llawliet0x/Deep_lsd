```python
import numpy as np
import os

# 定义 .npz 文件所在的目录
npz_directory = '/home/hh/Documents/DeepLSD/my_experiment'

# 列出目录中的所有 .npz 文件
npz_files = [f for f in os.listdir(npz_directory) if f.endswith('.npz')]

# 查看每个 .npz 文件的内容
for npz_file in npz_files:
    npz_path = os.path.join(npz_directory, npz_file)
    data = np.load(npz_path)
    
    print(f"Contents of {npz_file}:")
    for key in data.files:
        print(f" - {key}: shape {data[key].shape}")
```

```python
import numpy as np

# 加载 .npz 文件
file_path = '/home/hh/Documents/DeepLSD/my_experiment/detection_wireframe/deeplsd.npz'
data = np.load(file_path)

# 查看文件中的数组键名
print("文件中的数组键名：", data.files)

# 逐个查看每个数组的内容和形状
for key in data.files:
    print(f"键名: {key}, 数组形状: {data[key].shape}")
    print(data[key])  # 打印数组内容
```
